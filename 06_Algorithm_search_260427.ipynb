```python
#recursive call를 활용한 binary sarch
def binary_search_recursive(A, key, low, high): #binary_search함수 지정
    if(low>high):
        return -1 #탐색 항목이 없는 경우

    middle = (low + high)//2 #중앙값을 찾기 위하여

    if (key == A[middle]): #중앙값과 같으면 index를 return
        return middle
    elif (key < A[middle]): #탐색 영역을 왼쪽으로 설정한다
        return binary_search_recursive(A, key, low, middle-1)
    else: #탐색 영역을 오른쪽으로 설정한다
        return binary_search_recursive(A, key, middle + 1, high)
```

```python
#while loop를 활용한 binary search
def binary_search(A, key):#binary_search 함수 지정
    low = 0 #왼쪽을 0으로 두어 시작
    high = len(A) -1 #자리값이 0부터 시작이니 자기자신을 빼야 해서 -1을 함
    while low <= high:
        middle = (low + high)//2 #리스트 개수가 짝수면 중앙값을 찾기 위해 2로 나눈다
        if A[middle] == key:
            return middle #중앙값이 키값이면 다시 반환
        elif A[middle] > key:
            high = middle -1 #중앙값이 키값보다 크면 왼쪽에서 찾기 위해 -1
        else:
            low = middle + 1 #중앙값이 키값보다 작으면 오른쪽에서 찾기 위해 +1
    return -1 #whlie문을 끝내기 위해 -1을 반환하여 끝
```

```python
data = [2, 6, 11, 13, 18, 20, 22, 27, 29, 30, 34, 38, 41, 42, 45, 47]
num = int(input('찾을 숫자 :'))
index = binary_search_recursive(data, num, 0, len(data))
if index == -1:
    print('없는 숫자입니다.')
else:
    print(f'>>> 숫자 {num}은 index { index}에 있습니다')

index = binary_search(data, num)
if index == -1:
    print('없는 숫자입니다.')
else:
    print(f'>>> 숫자 {num}은 index { index}에 있습니다')
```
